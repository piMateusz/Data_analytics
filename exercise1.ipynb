```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("Data1.csv")
df.head()
```

```python
# set first column as index
df = df.set_index(df.columns[0])
df.head()
```

```python
# Plot all columns as time series.
df.plot(subplots=True, figsize=(15, 10));
```
